Imports

```python
import matplotlib.pyplot as plt
from aeg import run_simulation
```

Run Simulations

```python
# --- same seed for fair comparison ---
t5_a, n5_a, x5_a, y5_a = run_simulation(5, seed=10, mutation_rate1=0.02, mutation_rate2=0.03)
t6_a, n6_a, x6_a, y6_a = run_simulation(6, seed=10, mutation_rate1=0.02, mutation_rate2=0.03)

t5_n, n5_n, x5_n, y5_n = run_simulation(5, seed=10, mutation_rate1=0.0, mutation_rate2=0.0)
t6_n, n6_n, x6_n, y6_n = run_simulation(6, seed=10, mutation_rate1=0.0, mutation_rate2=0.0)
```

Plotting Section

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 16))

# Font sizes
label_fontsize = 16
title_fontsize = 16
tick_fontsize = 14

# Column titles
axes[0,0].set_title("Non-adaptive (u = 0)", fontsize=title_fontsize)
axes[0,1].set_title("Adaptive (u > 0)", fontsize=title_fontsize)

# -------- s = 5 densities --------
for i in range(n5_n.shape[0]):
    axes[0,0].plot(t5_n, n5_n[i])
    axes[0,1].plot(t5_a, n5_a[i])

axes[0,0].set_ylabel("Density (s = 5)", fontsize=label_fontsize)
axes[0,0].tick_params(axis='both', labelsize=tick_fontsize)
axes[0,1].tick_params(axis='both', labelsize=tick_fontsize)

# -------- s = 5 traits --------
for i in range(x5_n.shape[0]):
    # Non-adaptive (traits static)
    axes[1,0].scatter(x5_n[i,0], y5_n[i,0], color=f"C{i}", marker='o', s=80)

    # Adaptive (full trajectories)
    axes[1,1].plot(x5_a[i], y5_a[i])
    axes[1,1].scatter(x5_a[i,0], y5_a[i,0], color=f"C{i}", marker='o', s=80)
    axes[1,1].scatter(x5_a[i,-1], y5_a[i,-1], color=f"C{i}", marker='x', s=80)

axes[1,0].set_ylabel("Traits (s = 5)", fontsize=label_fontsize)
axes[1,0].set_xlabel("Trait x", fontsize=label_fontsize)
axes[1,1].set_xlabel("Trait x", fontsize=label_fontsize)

# -------- s = 6 densities --------
for i in range(n6_n.shape[0]):
    axes[2,0].plot(t6_n, n6_n[i])
    axes[2,1].plot(t6_a, n6_a[i])

axes[2,0].set_ylabel("Density (s = 6)", fontsize=label_fontsize)
axes[2,0].tick_params(axis='both', labelsize=tick_fontsize)
axes[2,1].tick_params(axis='both', labelsize=tick_fontsize)

# -------- s = 6 traits --------
for i in range(x6_n.shape[0]):
    # Non-adaptive
    axes[3,0].scatter(x6_n[i,0], y6_n[i,0], color=f"C{i}", marker='o', s=80)

    # Adaptive
    axes[3,1].plot(x6_a[i], y6_a[i])
    axes[3,1].scatter(x6_a[i,0], y6_a[i,0], color=f"C{i}", marker='o', s=80)
    axes[3,1].scatter(x6_a[i,-1], y6_a[i,-1], color=f"C{i}", marker='x', s=80)

axes[3,0].set_ylabel("Traits (s = 6)", fontsize=label_fontsize)
axes[3,0].set_xlabel("Trait x", fontsize=label_fontsize)
axes[3,1].set_xlabel("Trait x", fontsize=label_fontsize)

# Y-labels for trait plots
axes[1,0].set_ylabel("Trait y", fontsize=label_fontsize)
axes[1,1].set_ylabel("Trait y", fontsize=label_fontsize)
axes[3,0].set_ylabel("Trait y", fontsize=label_fontsize)
axes[3,1].set_ylabel("Trait y", fontsize=label_fontsize)

plt.tight_layout() 
#plt.savefig("aeg_simulation.tiff", dpi=300, bbox_inches='tight') 
plt.show()
```
