Imports

```python
import matplotlib.pyplot as plt
from aeg import run_simulation
```

Run Example

```python
t5, n5, x5, y5 = run_simulation(5, seed=10)
t6, n6, x6, y6 = run_simulation(6, seed=10)
```

Plotting Section

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Set general font sizes
label_fontsize = 16
title_fontsize = 16
tick_fontsize = 14
legend_fontsize = 14

# --- s = 5 densities ---
for i in range(n5.shape[0]):
    axes[0,0].plot(t5, n5[i])
axes[0,0].set_title("Population dynamics (s = 5)", fontsize=title_fontsize)
axes[0,0].set_xlabel("Time", fontsize=label_fontsize)
axes[0,0].set_ylabel("Density", fontsize=label_fontsize)
axes[0,0].tick_params(axis='both', labelsize=tick_fontsize)

# --- s = 5 traits ---
for i in range(x5.shape[0]):
    axes[0,1].plot(x5[i], y5[i])
    axes[0,1].scatter(x5[i,0], y5[i,0], color=f"C{i}", marker='o', s=100)  # start
    axes[0,1].scatter(x5[i,-1], y5[i,-1], color=f"C{i}", marker='x', s=100) # end
axes[0,1].set_title("Evolutionary trajectories (s = 5)", fontsize=title_fontsize)
axes[0,1].set_xlabel("Trait x", fontsize=label_fontsize)
axes[0,1].set_ylabel("Trait y", fontsize=label_fontsize)
axes[0,1].tick_params(axis='both', labelsize=tick_fontsize)

# --- s = 6 densities ---
for i in range(n6.shape[0]):
    axes[1,0].plot(t6, n6[i])
axes[1,0].set_title("Population dynamics (s = 6)", fontsize=title_fontsize)
axes[1,0].set_xlabel("Time", fontsize=label_fontsize)
axes[1,0].set_ylabel("Density", fontsize=label_fontsize)
axes[1,0].tick_params(axis='both', labelsize=tick_fontsize)

# --- s = 6 traits ---
for i in range(x6.shape[0]):
    axes[1,1].plot(x6[i], y6[i])
    axes[1,1].scatter(x6[i,0], y6[i,0], color=f"C{i}", marker='o', s=100)  # start
    axes[1,1].scatter(x6[i,-1], y6[i,-1], color=f"C{i}", marker='x', s=100) # end
axes[1,1].set_title("Evolutionary trajectories (s = 6)", fontsize=title_fontsize)
axes[1,1].set_xlabel("Trait x", fontsize=label_fontsize)
axes[1,1].set_ylabel("Trait y", fontsize=label_fontsize)
axes[1,1].tick_params(axis='both', labelsize=tick_fontsize)

plt.tight_layout()
#plt.savefig("amg_simulation.tiff", dpi=300, bbox_inches='tight')
plt.show()
```
